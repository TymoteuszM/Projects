```python
#importowanie bibliotek, wczytanie danych
import pandas as pd
import numpy as np
import seaborn as sns
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
#gender_data = pd.read_csv("gender_submission.csv")
```

```python
#usuniÄ™cie kolumny survived z train data, poÅ‚Ä…czenie train data i test data w celu wykorzystnaia informacji o parch i sibsp
#wykorzystanie samego nazwiska (wyciecie imienia,tytulu itd), uzupelnienei kabin i zamiana na pierwsza litere kabiny 
survived=train_data["Survived"]

train_data = train_data.drop(["Survived"], axis = 1)
df=pd.concat([train_data,test_data])
df=df.reset_index()
df["Name"] = df["Name"].str.split(pat = ",")
df["Name"] = df["Name"].apply(lambda x: x[0])
df["Cabin"] = df["Cabin"].fillna("0")
df["Cabin_type"] = df["Cabin"].apply(lambda x: x[0])
```

```python
#Zapisanie przedzialu wartosci age i sibsp , zamiaan plci na 0-1, usuniecie biletu, przeskalowanie zmiennych na przedziaÅ‚ 0-1
# one hot encoding na Embarked oraz Cabin_type
min_age=min(df["Age"])
max_age=max(df["Age"])

min_sibsp=min(df["SibSp"])
max_sibsp=max(df["SibSp"])
df["Sex"] = (df["Sex"] == "male").astype(int)
df1 = pd.get_dummies(df["Embarked"], drop_first = True) # one hot encoding
df2 = pd.get_dummies(df["Cabin_type"], drop_first = True) # one hot encoding
df = pd.concat([df, df1,df2], axis = 1)
```

```python
#UsuniÄ™cie kolumn z nazwiskiem, passengerid, kabinom, biletem, typem kabiny oraz miejscem wejÅ›cia na pokÅ‚ad
name=df["Name"]
ticket=df["Ticket"]
df = df.drop(["PassengerId", "Cabin",  "Ticket","Cabin_type","Embarked","index"], axis = 1)
df=df.drop(["Name"],axis=1)
#Przeskalowanie danych przy pomocy MinMax Scalera w celu uzupeÅ‚nienia brakujÄ…cych danych
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
```

```python
df.head()
```

```python
#Uzupelnienie wartosci bez zmiennych przy pomocy k najblizszych sÄ…siadÃ³w

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 5)
df = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
```

```python
#PowrÃ³t do wartoÅ›ci age i sibsp nieprzeskalowanych
df["Age"] = df["Age"] * (max_age - min_age) + min_age
df["SibSp"] = df["SibSp"] * (max_sibsp - min_sibsp) + min_sibsp
```

```python
#Dodanie kolumny name 
df["Name"]=name
df["Ticket"]=ticket
```

```python
#Wyliczenie liczby wystÄ™powania osÃ³b o tych samych nazwiskach oraz osÃ³b z tym samym biletem, przyporzÄ…dkowanie poszczegÃ³lnym 
#osobom liczbe osob o tym samym nazwisku oraz o tym samym bilecie

xx=df["Name"].value_counts()
yy=df["Ticket"].value_counts()
df["Same_Ticket"]=df["Ticket"].apply(lambda x : yy[x])
df["Same_Name"]=df["Name"].apply(lambda x : xx[x])
```

```python
#Wprowadzenie dodatkowych zmiennych - samotny wilk- osoba podrÃ³Å¼ujÄ…ca samodzielnie, oraz Alon_mar - samotne maÅ‚Å¼eÅ„stwo

df["Wolf"]=df["SibSp"]+df["Parch"]
df["Wolf"]=df["Wolf"].apply(lambda  x : np.abs(int(x>0)-1))

df["Alon_mar_1"]=df["SibSp"].apply(lambda x : (int(x)==1))
df["Alon_mar_2"]=df["Parch"].apply(lambda x : x==0)
df["Alon_mar"]=df["Alon_mar_1"]&df["Alon_mar_2"]
df["Alon_mar"]=df["Alon_mar"].astype(int)


#UsuniÄ™cie niepotrzebnych kolumn
df = df.drop(["Name", "Ticket",  "Alon_mar_1","Alon_mar_2"], axis = 1)
#Ponowne przeskalowanie danych
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)


```

```python
df.head()
```

```python
#PodziaÅ‚ zbioru danych na dane treningowe oraz testowe
train_data=df.iloc[0:891,]
test_data=df.iloc[891:,]

xtrain=train_data
xtest=test_data
ytrain=survived
```

```python
#Regresja logistyczna dla danych uzupelnionych z sipsp i parch



from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(xtrain,ytrain)

predictions=log_reg.predict(xtest)
```

```python
e={'PassengerId':test_data.index+1,'Survived' :(predictions.astype(int))}
data_submission=pd.DataFrame(data=e)
data_submission.to_csv(r'data_submission_titanic_tm_logistic_regression2.csv',index=False)
```

```python
# metoda k najblizszych sasiadow do wyznaczenia prawidlowosci modelu

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(train_data,survived,test_size=0.2,random_state=2)
X = xtrain
y = ytrain
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)

ypredict=neigh.predict(xtest)

ypredict2=neigh.predict(test_data)

from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,ypredict))
```

```python
#Zapisanie wynikÃ³w w pliku csv

d={'PassengerId':test_data.index+1,'Survived' :(ypredict2.astype(int))}
data_submission=pd.DataFrame(data=d)
data_submission.to_csv(r'data_submission_knn_tm.csv',index=False)
```

```python
#Importowanie bibliotek do XGboosta oraz do optymalizacji parametrÃ³w
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

```

```python
#Optymalizacja parametrÃ³w 
parameters= {
    'use_label_encoder':[False],
    'eta': [0.01,0.0175,0.025,0.0325,0.05,0.1,0.15,],
    'max_depth':[3,4,5,6,7],
    'subsample':[0.35,0.4,0.45,0.5]      
}

from sklearn.model_selection import cross_val_score
clf=XGBClassifier(use_label_encoder=False,eval_metric='mlogloss')

clf = GridSearchCV(clf, parameters)
clf.fit(xtrain ,ytrain)

#sorted(clf.cv_results_.keys())

print(clf.best_params_,clf.best_score_)
```

```python
#Obliczenia dla najlepszych parametrÃ³w XGBClassifiera, zapisanie danych w pliku csv
xtrain=train_data
xtest=test_data
ytrain=survived


clf2=XGBClassifier(use_label_encoder=False,eval_metric='mlogloss',eta=0.1,subsample=0.35,max_depth=6)
clf2.fit(xtrain,ytrain)

predictions=clf2.predict(xtest)

e={'PassengerId':test_data.index+1,'Survived' :(predictions.astype(int))}
data_submission=pd.DataFrame(data=e)
data_submission.to_csv(r'data_submission_tm_XGBOOST_2.csv',index=False)
```
