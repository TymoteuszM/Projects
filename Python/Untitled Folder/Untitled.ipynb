```python
def triangle(nazwa):
    data = pd.read_csv(nazwa, sep=" ", header=None)
    data=data[list(range(len(data),0,-1))]
    data=data.replace(np.nan,0)
    data.columns = list(range(0,len(data)))
    maksimum=0.0
    sciezka=[]
    print(data)
    for i in range(len(data)):
        for j in range(len(data)):
            for k in range(len(data)):
                for l in range(len(data)):
                    for m in range(len(data)):
                        for n in range(len(data)):
                            suma=data[i][j]+data[i][k]+data[i][l]+data[i][m]+data[i][n]
                            print(suma)
                            if(maksimum<suma):
                                maksimum=suma
                                sciezka=[data[i][j],data[i][k],data[i][l],data[i][m],data[i][n]]
    print(maksimum)
    print(sciezka)
```
