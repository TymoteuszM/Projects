```python
import pandas as pd
import numpy as np
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input,Output
import plotly.graph_objects as go


```

```python

# Load Data
df=pd.read_csv('athlete_events.csv')
countries=pd.read_csv('country_definitions.csv')
df=df.loc[df['Season'] == 'Summer']
medals=pd.read_csv('Summer_olympic_Medals.csv')
kraje=np.sort(medals["Country_Name"].unique())
```

```python
# clean data
df[['Age','Height','Weight']] = df[['Age','Height','Weight']].fillna(0)

df.Medal = df.Medal.fillna('None')

df.Age = df.Age.astype(int)

top_countries = df.Team.value_counts().sort_values(ascending=False).head(10)

lata=np.sort(df["Year"].unique())

```

```python


#aplikacja
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Letnie Igrzyska Olimpijskie 1896-2016",style={'text-align':'center'}),
        
    dcc.Slider(id="select_year",
        min=lata[0],
        included=False,
        max=lata[-1],
        value=lata[-2],
        marks={int(i) : str(i) for i in lata},
    
    ),
    html.Div(id='output_containerw',style={'text-align':'center'}),
    dcc.Graph(id='world_chart'),

    html.Div(id='output_container_g',style={'text-align':'center'}),
    dcc.Graph(id='world_chart_g'),
    html.Div(id='output_container_s',style={'text-align':'center'}),
    dcc.Graph(id='world_chart_s'),
    html.Div(id='output_container_b',style={'text-align':'center'}),
    dcc.Graph(id='world_chart_b'),
    
    
    html.Div(id='output_container0',style={'text-align':'center'}),
    dcc.Graph(id='bar_chart'),
    
    html.Div(id='output_container1'),
    html.Br(),
    
    dcc.Graph(id='age_hist'),
    html.Div(id='output_container2'),
    dcc.Graph(id='hist_height'),
    html.Div(id='output_container3'),
    dcc.Graph(id='hist_weight'),
    html.Div(id='output_container4'),
    dcc.Graph(id='pie_gender'),
                 
             
     html.Div([
        html.H1(children='Liczba medali na przestrzeni lat dla wybranego kraju '),

         dcc.Dropdown(id="select_country",
                    options=[
                        {'label':str(i), 'value':str(i)} for i in kraje
                      ],
                      value=kraje[0],
                    searchable=True,
                    placeholder="Wybierz nazwę kraju",
                      ),
        html.Div(id='output_container_line'),
        dcc.Graph(
            id='graph_line'
        ),  
    ]),


                      
 html.Div([
        html.H1(children='Porównanie liczby medali na przestrzeni lat '),
        dcc.Graph(id='graph_line_two'), 
     ])
    
    
])
    
    
               
@app.callback(
    [
    Output(component_id='output_containerw',component_property='children'),
    Output(component_id='world_chart',component_property='figure'),
    Output(component_id='output_container_g',component_property='children'),
    Output(component_id='world_chart_g',component_property='figure'),
    Output(component_id='output_container_s',component_property='children'),
    Output(component_id='world_chart_s',component_property='figure'),
    Output(component_id='output_container_b',component_property='children'),
    Output(component_id='world_chart_b',component_property='figure'),
    Output(component_id='output_container0',component_property='children'),
    Output(component_id='bar_chart',component_property='figure'),
    Output(component_id='output_container1',component_property='children'),
    Output(component_id='age_hist',component_property='figure'),
    Output(component_id='output_container2',component_property='children'),
    Output(component_id='hist_height',component_property='figure'),
    Output(component_id='output_container3',component_property='children'),
    Output(component_id='hist_weight',component_property='figure'),
    Output(component_id='output_container4',component_property='children'),
    Output(component_id='pie_gender',component_property='figure'),
    Output(component_id='output_container_line',component_property='children'),
    Output(component_id='graph_line',component_property='figure'),
    #Output(component_id='output_container_line_two',component_property='children'),
    Output(component_id='graph_line_two',component_property='figure')
    ],
    [Input(component_id='select_year',component_property='value'),
    Input(component_id='select_country',component_property='value')
    ]
)
def update_graph(option_slctd,option_slctd2):
    print(option_slctd)
    print(type(option_slctd))
    print(option_slctd2)
    print(type(option_slctd2))
    
    container_g="Rozkład złotych medali dla roku:{}".format(option_slctd)
    container_s="Rozkład srebrnych medali dla roku:{}".format(option_slctd)
    container_b="Rozkład brązowych medali dla roku:{}".format(option_slctd)
    containerw="Rozkład medali dla roku:{}".format(option_slctd)
    container0="Rozkład medali dla roku:{}".format(option_slctd)
    container1 = "Histogram wieku dla roku:{}".format(option_slctd)
    container2 ="Histogram wzrostu dla roku:{}".format(option_slctd)
    container3= "Histogram wagi dla roku:{}".format(option_slctd)
    container4= "Gender distribution dla roku:{}".format(option_slctd)
    container_line ="Liczba medali na przestrzeni lat dla:{}".format(option_slctd2)
    #container_line_two ="Porównanie liczby medali na przestrzeni lat dla:{}".format(option_slctd2)
    global df
    dff=df.copy()
    dff=dff[dff["Year"]==option_slctd]
    gender_counts=dff.Sex.value_counts()
    dff1=dff[dff["Age"]!=0]
    dff2=dff[dff["Height"]!=0]
    dff3=dff[dff["Weight"]!=0]
    fig1=px.histogram(dff1,x="Age",color=dff1["Sex"],barmode='overlay')
    fig2=px.histogram(dff2,x="Height",color=dff2["Sex"],barmode='overlay')
    fig3=px.histogram(dff3,x="Weight",color=dff3["Sex"],barmode='overlay')
    fig4=px.pie(gender_counts, values=gender_counts,names=gender_counts.index)
    
    medals_probe=medals[medals["Year"]==option_slctd]
    world=px.choropleth(medals_probe,locations ="Country_Name" ,
               locationmode = 'country names',color="Total",
                       color_continuous_scale=px.colors.sequential.Greens)
    world_g=px.choropleth(medals_probe,locations ="Country_Name" ,
                   locationmode = 'country names',color="Gold",
                         color_continuous_scale=px.colors.sequential.YlOrBr)
    world_s=px.choropleth(medals_probe,locations ="Country_Name" ,
                   locationmode = 'country names',color="Silver",
                         color_continuous_scale=px.colors.sequential.Greys)    
    world_b=px.choropleth(medals_probe,locations ="Country_Name" ,
                   locationmode = 'country names',color="Bronze",
                         color_continuous_scale=px.colors.sequential.amp)    
    
    medals_probe = medals_probe[['Country_Name','Bronze','Silver','Gold','Total']].sort_values(by=['Total','Gold','Silver'], ascending=[False,False,False]).head(20)
    
    

    
    
    
    fig0 = go.Figure(
        data = [
        # data for gold medals 
        go.Bar(
            name='Gold', 
            y=medals_probe['Country_Name'], 
            x=medals_probe['Gold'],
            orientation='h', 
            marker=dict(color='gold', line=dict(color='black', width=1))
        ),
        # data for silver medals 
        go.Bar(
            name='Silver', 
            y=medals_probe['Country_Name'], 
            x=medals_probe['Silver'],
            orientation='h', 
            marker=dict(color='silver', line=dict(color='black', width=1))
        ),
        # data for bronze medals 
        go.Bar(
            name='Bronze', 
            y=medals_probe['Country_Name'], 
            x=medals_probe['Bronze'],
            orientation='h', 
            marker=dict(color='chocolate', line=dict(color='black', width=1))
        )
    ])
    fig0.update_layout(
    margin = dict(pad=5),
    title = dict(
        text=f'Medal distribution of Top 20 countries ranked',
        font=dict(family='Rockwell', size=20)
    ),
    font = dict(family='Verdana', size=12),
    barmode='stack',
    height = 800, 
    width = 750, 
    template = 'plotly_white', 
    legend = dict(title='Medal')
)
    medals_two=medals
    medals_total_two=list(medals["Gold"]+medals["Silver"]+medals["Bronze"])
    medals_two["Total"]=medals_total_two
    medals_two.sort_values(["Year","Country_Name"],ascending=[True,True])
    fig_line_two=px.line(medals_two,x="Year",y="Total",
                     color="Country_Name",markers=True
                    )
    
    
    
    medals2=medals[medals["Country_Name"]==option_slctd2]
    medals_total=list(medals2["Gold"]+medals2["Silver"]+medals2["Bronze"])
    medals2["Total"]=medals_total
    
    fig_line=px.line(medals2,x="Year",y=["Gold","Silver","Bronze","Total"],
                     color_discrete_sequence =['gold','silver','saddlebrown','red'],markers=True
                    )

    return (containerw, world, 
            container_g, world_g,
            container_s, world_s,
            container_b, world_b,
            container0, fig0,
            container1, fig1,
            container2, fig2,
            container3, fig3,
            container4, fig4,
            container_line, fig_line,
            fig_line_two
           )

if __name__ == '__main__':
    app.run_server(debug=True)
```
