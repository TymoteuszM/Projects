```python
x=[0]
y=[0]
import matplotlib.pyplot as plt
import random
n=100000
for i in range(n):
    kierunek=random.randint(1,3)
    if kierunek==1:
        x.append((x[i]+0)/2)
        y.append((y[i]+0)/2)
    if kierunek ==2:
        x.append((x[i]+1/2)/2)
        y.append((y[i]+(3**(1/2)/3))/2)       
    if kierunek ==3:
        x.append((x[i]+1)/2)
        y.append((y[i]+0)/2)
plt.scatter(x,y,s=0.5)
plt.show()
```
